```python
#Questions 1-4 
import pandas as pd
import statsmodels.api as sm

# Dataset
url = "https://raw.githubusercontent.com/AlR0d/courses/main/4421/kiel.csv"
data = pd.read_csv(url)

# Step 1: Define constants and model variables for each year
data['constant'] = 1  # Adding a constant term for intercept

# Separate data for 1981 and 1978
data_1981 = data[data['y81'] == 1]
data_1978 = data[data['y81'] == 0]

# Step 2: Estimate the models for each year
# Model 1: Regression for 1981 (Equation 1)
model_1981 = sm.OLS(data_1981['rprice'], data_1981[['constant', 'nearinc']]).fit()

# Model 2: Regression for 1978 (Equation 2)
model_1978 = sm.OLS(data_1978['rprice'], data_1978[['constant', 'nearinc']]).fit()

# key results for Model 1 (1981)
print("Model 1 (1981):")
print("Coefficients:\n", model_1981.params)
print("P-Values:\n", model_1981.pvalues)

#  key results for Model 2 (1978)
print("\nModel 2 (1978):")
print("Coefficients:\n", model_1978.params)
print("P-Values:\n", model_1978.pvalues)

# Step 3: Difference-in-difference estimator
# δ = β1 - α1
diff_in_diff_estimator = model_1981.params['nearinc'] - model_1978.params['nearinc']
print(f"\nDifference-in-Difference Estimator (δ): {diff_in_diff_estimator}")

# Step 4: Hypothesis testing with Equation 3 (with y81 and interaction term)
data['nearinc_y81'] = data['nearinc'] * data['y81']  # Interaction term
model_3 = sm.OLS(data['rprice'], data[['constant', 'nearinc', 'y81', 'nearinc_y81']]).fit()

# Display key results for Model 3
print("\nModel 3 (with Interaction Term):")
print("Coefficients:\n", model_3.params)
print("P-Values:\n", model_3.pvalues)

# Step 5: Add additional explanatory variables to model
additional_vars = ['age', 'agesq', 'rooms', 'baths', 'land', 'area']
model_expanded = sm.OLS(data['rprice'], data[['constant', 'nearinc', 'y81', 'nearinc_y81'] + additional_vars]).fit()

# Display key results for the expanded model
print("\nExpanded Model with Additional Variables:")
print("Coefficients:\n", model_expanded.params)
print("P-Values:\n", model_expanded.pvalues)
```

Question #4 
The topic of the research paper was to be able to quantify the impact of technical change on economic output while disregarding the effects of labor and capital inputs. For Solow’s approach, he used the aggregate production function of the form: 
Q=F(K,L,t), where Q is output, K is capital, L is labor and t is time. Solow assumes that technical change shifts the production function without changing the marginal rate of substitution between labor and capital. He then derives the growth accounting equation and calculates the “Solow Residual” which is a measure of total factor productivity. And this residual represents the portion of output growth unexplained by growth in capital and labor inputs. The conclusion of this paper was that economic growth is heavily driven by technical change rather than the sum of capital and labor, the Solow residual was able to show productivity growth from factors such as innovation, education and better practices. In addition, the Hicks-Neutral assumption is also discussed to imply that technology affects output without changing the contribution of labor and capital.

```python
##Question 6 
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.iolib.summary2 import summary_col

# Load the data (example with simulated data, replace this with actual data loading if needed)
data_dict = {
    "year": [1978, 1978, 1978, 1978, 1978, 1978, 1978, 1978, 1978, 1978, 1978, 1978, 1978, 1978, 1978, 1978, 1978, 1978, 1978, 1978],
    "q": [60000, 40000, 34000, 63900, 44000, 46000, 56000, 38500, 60500, 55000, 39000, 41000, 50900, 52000, 49000, 80000, 50000, 59000, 42000, 71500],
    "k": [10700, 11000, 11500, 11900, 12100, 10000, 11700, 10200, 10500, 11000, 8600, 12200, 12400, 16800, 17200, 9900, 10000, 10000, 10600, 11700]
}

# Create DataFrame
data = pd.DataFrame(data_dict)

# Define the dependent and independent variables for each model
data['ln_q'] = np.log(data['q'])
data['ln_k'] = np.log(data['k'])
data['inverse_k'] = 1 / data['k']

# Model 4a: q = α + βk
X4a = sm.add_constant(data['k'])
model_4a = sm.OLS(data['q'], X4a).fit()

# Model 4b: q = α + β log(k)
X4b = sm.add_constant(data['ln_k'])
model_4b = sm.OLS(data['q'], X4b).fit()

# Model 4c: q = α - β / k
X4c = sm.add_constant(data['inverse_k'])
model_4c = sm.OLS(data['q'], X4c).fit()

# Model 4d: log(q) = α + β log(k) (Cobb-Douglas form)
X4d = sm.add_constant(data['ln_k'])
model_4d = sm.OLS(data['ln_q'], X4d).fit()

# Model 4e: log(q) = α - β / k
X4e = sm.add_constant(data['inverse_k'])
model_4e = sm.OLS(data['ln_q'], X4e).fit()

# Collect all models for summary table
models = [model_4a, model_4b, model_4c, model_4d, model_4e]
model_names = ["Model 4a (q = α + βk)", 
               "Model 4b (q = α + β log(k))", 
               "Model 4c (q = α - β / k)", 
               "Model 4d (log(q) = α + β log(k))", 
               "Model 4e (log(q) = α - β / k)"]

# Display summaries of all models in a neat table
model_summaries = summary_col(models, stars=True, model_names=model_names,
                              info_dict={'R-squared': lambda x: f"{x.rsquared:.3f}",
                                         'No. observations': lambda x: f"{int(x.nobs)}"})

print("\nModel Summary Table:")
print(model_summaries)
```

```python
Question #11 
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy.stats import f

# Define a function to perform the Chow test
def chow_test(data, y, x, split_variable):
    """
    Perform Chow test for structural breaks.
    
    Parameters:
        data (DataFrame): The dataset.
        y (str): Dependent variable.
        x (list): Independent variables.
        split_variable (str): The variable to split the data.
        
    Returns:
        F-statistic and p-value.
    """
    # Ensure no missing or invalid values
    data = data.dropna(subset=[y] + x + [split_variable])
    
    # Split the data into base and treatment groups
    data_base = data[data[split_variable] == 0]
    data_treatment = data[data[split_variable] == 1]
    
    # Fit models for each group
    model_base = sm.OLS(data_base[y], sm.add_constant(data_base[x])).fit()
    model_treatment = sm.OLS(data_treatment[y], sm.add_constant(data_treatment[x])).fit()
    
    # Combine the groups into a single model
    model_combined = sm.OLS(data[y], sm.add_constant(data[x + [split_variable]])).fit()
    
    # Calculate residual sum of squares (RSS)
    rss_base = np.sum(model_base.resid ** 2)
    rss_treatment = np.sum(model_treatment.resid ** 2)
    rss_combined = np.sum(model_combined.resid ** 2)
    
    # Degrees of freedom
    n = len(data)
    k = len(model_combined.params)
    num = (rss_combined - (rss_base + rss_treatment)) / k
    denom = (rss_base + rss_treatment) / (n - 2 * k)
    
    # F-statistic and p-value
    f_stat = num / denom
    p_value = f.sf(f_stat, k, n - 2 * k)
    
    return f_stat, p_value

# Perform the Chow test for all five models
results = []
models = [
    {"y": "q", "x": ["k"], "name": "Model 4a"},
    {"y": "q", "x": ["ln_k"], "name": "Model 4b"},
    {"y": "q", "x": ["inverse_k"], "name": "Model 4c"},
    {"y": "ln_q", "x": ["ln_k"], "name": "Model 4d"},
    {"y": "ln_q", "x": ["inverse_k"], "name": "Model 4e"}
]

for model in models:
    f_stat, p_value = chow_test(data, model["y"], model["x"], "dummy_1943_1949")
    results.append({"Model": model["name"], "F-statistic": f_stat, "p-value": p_value})

# Display the results
results_df = pd.DataFrame(results)
print("Chow Test Results:")
print(results_df)
```
